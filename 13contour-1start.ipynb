```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_gray = cv.imread("img/kanban.png", cv.IMREAD_GRAYSCALE)
img = cv.imread("img/kanban.png")
# ret, threshold = cv.threshold(img_gray, 127, 255, 0)
img_gray = cv.medianBlur(img_gray, 33)
threshold = cv.adaptiveThreshold(img_gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
contours, hierarchy = cv.findContours(threshold, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# plt.figure(figsize=(20, 20))
plt.imshow(threshold, "gray")
```

```python
img_contour = img.copy()
cv.drawContours(img_contour, contours, -1, (0, 0, 255), 3)
img = cv.cvtColor(img, cv.COLOR_BGRA2RGBA)
img_contour = cv.cvtColor(img_contour, cv.COLOR_BGRA2RGBA)
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img_contour)
```

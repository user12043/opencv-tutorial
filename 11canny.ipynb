# CANNY EDGE DETECTION

https://docs.opencv.org/4.5.3/da/d22/tutorial_py_canny.html

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15, 15))
img = cv.imread("img/sudoku.jpg", 0)
img_edges = cv.Canny(img, 30, 100)

plt.subplot(1, 2, 1), plt.imshow(img, "gray")
plt.subplot(1, 2, 2), plt.imshow(img_edges, "gray")
```

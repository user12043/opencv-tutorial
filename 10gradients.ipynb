## 1. Sobel and Scharr Derivatives

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("img/sudoku.jpg", 0)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1), plt.imshow(img, "gray"), plt.xticks([])
plt.subplot(1, 3, 2), plt.imshow(cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5), "gray")
plt.subplot(1, 3, 3), plt.imshow(cv.Sobel(img, cv.CV_64F, 0, 1, ksize=5), "gray")
```

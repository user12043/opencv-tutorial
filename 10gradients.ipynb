## 1. Sobel and Scharr Derivatives

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("img/sudoku.jpg", 0)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1), plt.imshow(img, "gray"), plt.xticks([])
plt.subplot(1, 3, 2), plt.imshow(cv.Sobel(img, cv.CV_8UC1, 1, 0, ksize=5), "gray")
plt.subplot(1, 3, 3), plt.imshow(cv.Sobel(img, cv.CV_8UC1, 0, 1, ksize=5), "gray")
```

## 2. Laplacian Derivatives

```python
plt.figure(figsize=(15, 15))
laplacian = cv.Laplacian(img, cv.CV_16UC4)
plt.subplot(1, 2, 1), plt.imshow(laplacian, "gray")
# try some thresholding
ret, threshold = cv.threshold(cv.GaussianBlur(laplacian, (5, 5), 0), 0, 255, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)
plt.subplot(1, 2, 2), plt.imshow(threshold, "gray")
```

## Output Datatype Difference

```python
img = cv.imread("img/box.png")
plt.figure(figsize=(15, 15))
sobel8u = cv.Sobel(img, cv.CV_8U, 1, 0, ksize=5)

sobel64f = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5)
sobel64f_abs = np.absolute(sobel64f)
sobel8u_64f = np.uint8(sobel64f_abs)

plt.subplot(1, 3, 1), plt.imshow(img, "gray")
plt.subplot(1, 3, 2), plt.imshow(sobel8u, "gray")
plt.subplot(1, 3, 3), plt.imshow(sobel8u_64f, "gray")
```

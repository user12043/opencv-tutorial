# IMAGE PYRAMIDS

https://docs.opencv.org/4.5.3/dc/dff/tutorial_py_pyramids.html

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 15))
img = cv.imread("img/eagle_small.jpg")
img_low = cv.pyrDown(img)
img_high = cv.pyrUp(img_low)
plt.subplot(1, 3, 1), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 2), plt.imshow(cv.cvtColor(img_low, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 3), plt.imshow(cv.cvtColor(img_high, cv.COLOR_BGR2RGB))
```
